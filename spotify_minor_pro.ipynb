```python

# --- Data preprocessing ---
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load data (update path if running in Colab with Drive)
path = r'spotify.csv.csv'
df = pd.read_csv(path, low_memory=False)

# Basic info
print('Shape:', df.shape)
print('\nColumns:', list(df.columns)[:40])

# Drop duplicate rows if any
df = df.drop_duplicates().reset_index(drop=True)

# Show missing values summary
missing = df.isnull().sum().sort_values(ascending=False)
print('\nTop missing values:\n', missing.head(20))

# Heuristic: select numeric columns (float/int) for clustering and correlation
num_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
print('\nNumeric columns (detected):', num_cols)

# If there are no numeric columns detected (rare), try to coerce some columns
if len(num_cols) == 0:
    for c in df.columns:
        try:
            df[c] = pd.to_numeric(df[c])
        except:
            pass
    num_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
    print('After coercion, numeric columns:', num_cols)

# Fill missing numeric values with column median (simple approach)
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Standardize numeric features for modeling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[num_cols]) if len(num_cols)>0 else None

# Keep copies for later
processed = df.copy()
print('\nPreprocessing done.')

# EDA plots
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import scatter_matrix

# Use only first 8 numeric features for pairwise plots to keep it readable
if len(num_cols) == 0:
    print('No numeric columns to plot.')
else:
    plot_cols = num_cols[:8]
    print('Plotting columns:', plot_cols)
    # Histograms
    df[plot_cols].hist(figsize=(12,8))
    plt.suptitle('Histograms of selected numeric features')
    plt.show()

    # Boxplots
    fig, axs = plt.subplots(1, len(plot_cols), figsize=(3*len(plot_cols),4))
    for i, c in enumerate(plot_cols):
        axs[i].boxplot(df[c].dropna().values)
        axs[i].set_title(c)
    plt.suptitle('Boxplots')
    plt.show()

    # Scatter matrix (pairplot-like)
    scatter_matrix(df[plot_cols].sample(min(200, len(df))), figsize=(12,12))
    plt.suptitle('Scatter matrix (sampled)')
    plt.show()

# Correlation matrix
import matplotlib.pyplot as plt
import numpy as np

if len(num_cols) == 0:
    print('No numeric columns to compute correlation.')
else:
    corr = df[num_cols].corr()
    print('Correlation matrix shape:', corr.shape)
    # Plot using imshow
    fig, ax = plt.subplots(figsize=(10,8))
    cax = ax.imshow(corr.values, interpolation='nearest', aspect='auto')
    ax.set_xticks(range(len(corr.columns)))
    ax.set_yticks(range(len(corr.columns)))
    ax.set_xticklabels(corr.columns, rotation=90, fontsize=8)
    ax.set_yticklabels(corr.columns, fontsize=8)
    fig.colorbar(cax)
    plt.title('Correlation matrix (numeric features)')
    plt.tight_layout()
    plt.show()

# Clustering pipeline
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

if X_scaled is None:
    print('No numeric features available for clustering.')
else:
    # PCA for visualization
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_scaled)
    print('Explained variance by 2 PCAs:', pca.explained_variance_ratio_.sum())

    # Elbow method
    sse = []
    K = range(2,9)
    for k in K:
        km = KMeans(n_clusters=k, random_state=42, n_init=10)
        km.fit(X_scaled)
        sse.append(km.inertia_)
    plt.figure(figsize=(6,4))
    plt.plot(K, sse, marker='o')
    plt.xlabel('k')
    plt.ylabel('SSE (inertia)')
    plt.title('Elbow plot to choose k')
    plt.show()

    # Silhouette scores
    sil = []
    for k in K:
        km = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = km.fit_predict(X_scaled)
        sil.append(silhouette_score(X_scaled, labels))
    plt.figure(figsize=(6,4))
    plt.plot(K, sil, marker='o')
    plt.xlabel('k')
    plt.ylabel('Silhouette score')
    plt.title('Silhouette scores for different k')
    plt.show()

    # Choose k = 4 as a default (you can change after inspecting plots)
    k = 4
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = km.fit_predict(X_scaled)
    df['cluster'] = cluster_labels

    # Plot clusters on PCA plane
    plt.figure(figsize=(8,6))
    for cl in np.unique(cluster_labels):
        idx = cluster_labels == cl
        plt.scatter(X_pca[idx,0], X_pca[idx,1], label=f'Cluster {cl}', s=10)
    plt.legend()
    plt.title('KMeans clusters visualized on first 2 principal components')
    plt.xlabel('PC1'); plt.ylabel('PC2')
    plt.show()

    # If playlist_genre or playlist_name columns exist, show composition
    for col in ['playlist_genre','playlist_name']:
        if col in df.columns:
            comp = df.groupby(['cluster', col]).size().unstack(fill_value=0)
            print('\nCluster composition by', col)
            display(comp.head(20))

# KNN-based recommender (content-based)
from sklearn.neighbors import NearestNeighbors
import numpy as np

if X_scaled is None:
    print('No numeric features - recommender cannot be built.')
else:
    nn = NearestNeighbors(n_neighbors=6, algorithm='auto').fit(X_scaled)

    # A helper function: returns top_n similar songs by dataframe index
    def recommend_by_index(idx, top_n=5):
        if idx < 0 or idx >= len(df):
            raise IndexError('Index out of range.')
        dist, ind = nn.kneighbors([X_scaled[idx]], n_neighbors=top_n+1)
        # ind includes the query itself at position 0, so skip it
        rec_indices = [i for i in ind[0] if i != idx][:top_n]
        return df.iloc[rec_indices].copy().reset_index()

    # Example: recommend for first song (index 0)
    if len(df) > 0:
        print('Example recommendations for index 0:')
        try:
            recs = recommend_by_index(0, top_n=5)
            display(recs.head(10))
        except Exception as e:
            print('Could not show recommendations:', e)
```
